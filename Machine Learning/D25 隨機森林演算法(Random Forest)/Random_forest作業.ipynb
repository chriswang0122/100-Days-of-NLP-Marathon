### 作業目的: 了解何謂Bias-Variance Tradeoff

### Q1: 一般模型誤差可拆為三部分，這三部分的誤差分別是？

Answer: 
- Bias(偏差)
- Variance(方差)
- Irreducible Error(不可避免誤差)

### Q2: 請解釋何謂Bias-Variance Tradeoff

Answer:
$$\text{Total Error(整體誤差)} = \text{Bias}^2\text{(偏差)} + \text{Variance(方差)} + \text{Irreducible Error(不可避免誤差)}$$

- **Irreducible error** is "**noise**" that can't be reduced by algorithms.
- To get good predictions, need to find **a balance of bias and variance** that minimizes total error.
- The goal of any **supervised machine learning algorithm** is to achieve **low bias and low variance**.
- There is no escaping **the relationship between bias and variance** in machine learning.
	- *Increasing the bias* will *decrease the variance*. (**underfitting**)
	- *Increasing the variance* will *decrease the bias*. (**overfitting**)
